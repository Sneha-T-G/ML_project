Outlier Removal

```python
for i in ['Age','Salary','Joining Designation','Designation','Quarterly Rating','Total Business Value','Target']:
  plt.figure()
  plt.boxplot(New_data[i])
  plt.title
```

```python
def find_boundaries(df,variables):
  q1=df[variables].quantile(.25)
  q3=df[variables].quantile(.75)
  IQR=q3-q1
  lower_boundary=q1-1.5*IQR
  upper_boundary=q3+1.5*IQR
  return lower_boundary,upper_boundary
```

```python
lower_age,upper_age=find_boundaries(New_data,'Age')
print('upper age is', upper_age)
print('lower age is', lower_age)
```

```python
New_data['Age']=np.where(New_data['Age']>upper_age,upper_age,np.where(New_data['Age']<lower_age,lower_age,New_data['Age']))
```

```python
lower_salary,upper_salary=find_boundaries(New_data,'Salary')
print('upper salary is', upper_salary)
print('lower salary is', lower_salary)
```

```python
New_data['Salary']=np.where(New_data['Salary']>upper_salary,upper_salary,np.where(New_data['Salary']<lower_salary,lower_salary,New_data['Salary']))
```

```python
lower_join,upper_join=find_boundaries(New_data,'Joining Designation')
print('upper Joining Designation is', upper_join)
print('lower Joining Designation is', lower_join)
```

```python
New_data['Joining Designation']=np.where(New_data['Joining Designation']>upper_join,upper_join,np.where(New_data['Joining Designation']<lower_join,lower_join,New_data['Joining Designation']))
```

```python
lower_QR,upper_QR=find_boundaries(New_data,'Quarterly Rating')
print('upper Quarterly Rating is', upper_QR)
print('lower Quarterly Rating is', lower_QR)
```

```python
New_data['Quarterly Rating']=np.where(New_data['Quarterly Rating']>upper_QR,upper_QR,np.where(New_data['Quarterly Rating']<lower_QR,lower_QR,New_data['Quarterly Rating']))
```

```python
lower_TB,upper_TB=find_boundaries(New_data,'Total Business Value')
print('upper Total Business Value is', upper_TB)
print('lower Total Business Value is', lower_TB)
```

```python
New_data['Total Business Value']=np.where(New_data['Total Business Value']>upper_TB,upper_TB,np.where(New_data['Total Business Value']<lower_TB,lower_TB,New_data['Total Business Value']))
```

# **Scaling**

```python
x1 = x.drop(['Age','City','Gender'],axis=1)
x1
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max_scaler=MinMaxScaler(feature_range=(0,1))
x1 = min_max_scaler.fit_transform(x1)
x1
```

```python
x1=pd.DataFrame(x1,columns=['Education_Level','Salary','Joining Designation',	'Designation','Quarterly Rating','Total Business Value','reporting_year',	'reporting_month','reporting_day_of_week','joining_year','joining_month','joining_day_of_week'])
```

```python
x2=x.drop(['Education_Level','Salary','Joining Designation',	'Designation','Quarterly Rating','Total Business Value','reporting_year',	'reporting_month','reporting_day_of_week','joining_year','joining_month','joining_day_of_week'],axis=1)
x2
```

```python
x3=pd.concat([x2,x1],axis=1)
x3
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x3,y,random_state=42,test_size=.25)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logit_model=LogisticRegression()
logit_model.fit(x_train,y_train)
y_pred=logit_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
print('Accuracy is',accuracy_score(y_test,y_pred))
print('Precision is',precision_score(y_test,y_pred,average='macro'))
print('Recall is',recall_score(y_test,y_pred,average='macro'))
print('f1 score is',f1_score(y_test,y_pred,average='macro'))
print(confusion_matrix(y_test,y_pred))
```

## **Decision Tree**

```python
print('Accuracy is: ',accuracy_score(y_test,y_pred))
print('Precision is',precision_score(y_test,y_pred,average='macro'))
print('Recall is',recall_score(y_test,y_pred,average='macro'))
print('f1 score is',f1_score(y_test,y_pred,average='macro'))
print(confusion_matrix(y_test,y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
metric_k=[]
neighbors=np.arange(3,15)

for k in neighbors:
    classifier =KNeighborsClassifier(n_neighbors=k,metric='manhattan')
    classifier.fit(x_train,y_train)
    y_pred=classifier.predict(x_test)
    acc=accuracy_score(y_test,y_pred)
    metric_k.append(acc)
```

```python
plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k Value')
plt.ylabel('accuracy')
plt.grid()
```

```python
classifier=KNeighborsClassifier(n_neighbors=8,metric='manhattan')
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score
print('Accuracy is',accuracy_score(y_test,y_pred))
print('Precision is',precision_score(y_test,y_pred))
print('Recall is',recall_score(y_test,y_pred))
print('f1 score is',f1_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred)) 
```
